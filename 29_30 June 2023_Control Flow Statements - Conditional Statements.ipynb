```python
# Program to check whether a number is 'Palindrome number'   --- Second method -- Short method...

a = input()
if a[0::]==a[::-1]:    #  --- Reversing the STRING
    print("It is a Palindrome number")
else:
    print("It is not a Palindrome number")
print("END")
```

```python
# Program to check whether a number is 'Palindrome number'   ---  Third method -- Short method...

a = input()
if a==a[::-1]:    #  --- Reversing the STRING
    print("It is a Palindrome number")
else:
    print("It is not a Palindrome number")
print("END")
```
